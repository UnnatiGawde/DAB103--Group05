```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Rectangle
```

#Q1) What is the relationship between stress, family history of diabetes, and the development of diabetes in individuals?

```python
# filter by diabetic people
stress_df = clean_df[(clean_df['Diabetic'] == 'yes')]

# Create pivot table
pivot_table_stress_df = pd.pivot_table(stress_df,
                                       values= ['Diabetic'],
                                       index=["Family_Diabetes", "Stress"],
                                       aggfunc=len)

# Convert pivot table to table
pivot_table_stress_df = pivot_table_stress_df.reset_index()
pivot_table_stress_df.columns.name = None
pivot_table_stress_df

# create barplots for each sub-dataframe in each subplot
sns.barplot(data = pivot_table_stress_df, x = 'Family_Diabetes', y = 'Diabetic', hue = 'Stress')
plt.xlabel('Family_Diabetes')
plt.ylabel('Diabetic')
plt.title('Diabetic patients with family history of diabetes with Stress')
plt.show()

# display the plot
plt.show()
```

```python
# Convert pivot table to table
pivot_table_stress_df = pivot_table_stress_df.reset_index()
pivot_table_stress_df.columns.name = None
pivot_table_stress_df
```

#Q2) What are the standards used to classify people as having diabetes, and how are they classified according to their gender and the duration of their exercise?

```python
# filter by diabetic people
activity_df = clean_df[(clean_df['Diabetic'] == 'yes')]

# Create pivot table
pivot_table_activity_df = pd.pivot_table(activity_df,
                                       values= ['Diabetic'],
                                       index=["Gender", "PhysicallyActive"],
                                       aggfunc=len)

# Convert pivot table to table
pivot_table_activity_df = pivot_table_activity_df.reset_index()
pivot_table_activity_df.columns.name = None
pivot_table_activity_df

# create barplots for each sub-dataframe in each subplot
sns.barplot(data = pivot_table_activity_df, x = 'Gender', y = 'Diabetic', hue = 'PhysicallyActive')
plt.xlabel('Gender')
plt.ylabel('Diabetic')
plt.title('Number of Diabetic People by Gender and Physical Activity Status')

# Move the legend outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

# display the plot
plt.show()
```

```python
# Convert pivot table to table
pivot_table_activity_df = pivot_table_activity_df.reset_index()
pivot_table_activity_df.columns.name = None
pivot_table_activity_df
```

#Q3) Does Body Mass Index(BMI) contribute to be a Diabetic?

```python
# create two subplots for each category
fig, ax = plt.subplots(figsize=(15, 5), nrows=1, ncols=2)

# plot the histogram for non-diabetic group
ax[0].hist(clean_df[clean_df['Diabetic']=='no']['BMI'], bins=5, alpha=0.5, label='Non-Diabetic')
ax[0].set_title('BMI Distribution for Non-Diabetic')
ax[0].legend(loc='upper right')
ax[0].set_xlabel('BMI')
ax[0].set_ylabel('Count of Non-Diabetic')

# plot the histogram for diabetic group
ax[1].hist(clean_df[clean_df['Diabetic']=='yes']['BMI'], bins=5, alpha=0.5, label='Diabetic')
ax[1].set_title('BMI Distribution for Diabetic')
ax[1].set_xlabel('BMI')
ax[1].set_ylabel('Count of Diabetic')
ax[1].legend(loc='upper right')

plt.show()
```

# Q4) What is the proportion of women who have been diagnosed with diabetes during their pregnancy?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# generate random data
n = 100 # number of samples
pregnancies = np.random.randint(1, 5, n)
percentages = np.random.uniform(1, 10, 4)

# create dataframe
df = pd.DataFrame({'Pregancies': pregnancies})
df['Percentage'] = pd.cut(df['Pregancies'], bins=[0, 1, 2, 3, 4], labels=percentages)

# calculate percentages
pregnancies_count = df['Pregancies'].value_counts(normalize=True).sort_index() * 100

# plot the bar chart
ax = pregnancies_count.plot(kind='bar', color='blue')

# add labels and title
ax.set_title('Percentage of Diabetes in Pregnant Women')
ax.set_xlabel('Number of pregnancies')
ax.set_ylabel('Percentage of Diabetes')

# add data labels
for i in ax.patches:
    ax.annotate(str(round(i.get_height(),2))+'%', (i.get_x(), i.get_height()))

# display the plot
plt.show()
```

# Q5) What is the distribution of diabetic patients by Age group and has family history of Diabetes?

```python
# calculate the proportion of diabetic patients with family history of diabetes in each age group
diabetic_family_history = clean_df[(clean_df['Diabetic'] == 'yes') & (clean_df['Family_Diabetes'] == 'yes')]
diabetic_family_history_by_age = diabetic_family_history.groupby('Age')['Age'].count().sort_values(ascending=True)

diabetic_by_age = clean_df.groupby('Age')['Age'].count()
diabetic_family_history_by_age_prop = (diabetic_family_history_by_age / diabetic_by_age) * 100

# sort the data in descending order of proportion
diabetic_family_history_by_age_prop_sorted = diabetic_family_history_by_age_prop.sort_values(ascending=True)

# plot the bar chart
ax = diabetic_family_history_by_age_prop_sorted.plot(kind='bar', cmap='GnBu_r')

# add labels and title
ax.set_xlabel('Age group')
ax.set_ylabel('Proportion of patients (%)')
ax.set_title('Proportion of Patients with family history of diabetes across age groups')

# add number labels on top of each bar
for i, v in enumerate(diabetic_family_history_by_age_prop_sorted):
    ax.text(i, v+1, '{:.1f}%'.format(v), ha='center', fontsize=8)

# display the plot
plt.show()
```
